```python
longest_duo = df.loc[df["content_duration"].idxmax()]
shortest_duo = df.loc[df["content_duration"].idxmin()]


```

```python
#     
course_comparison = pd.DataFrame({
    'Course': [longest_duo['course_title'], shortest_duo['course_title']],
    'Duration': [longest_duo['content_duration'], shortest_duo['content_duration']]
})

#  Bar Chart 
plt.figure(figsize=(8, 6))
plt.bar(course_comparison['Course'], course_comparison['Duration'], color=['skyblue', 'red'])
plt.title('Longest & Shortest Course Duration', fontsize=14)
plt.xlabel('Course Title', fontsize=12)
plt.ylabel('Duration (hours)', fontsize=12)
plt.xticks(rotation=45, ha='right')  #    
plt.show()
```

```python
df['published_timestamp'] = pd.to_datetime(df["published_timestamp"])

df["month"] = df['published_timestamp'].dt.month
months_count = df['month'].value_counts().sort_index()
months_count
```

```python
def get_season(month):
    if month in [12,1,2]:
        return "Winter"
    elif month in [3,4,5]:
        return "Spring"
    elif month in [6,7,8]:
        return "Summer"
    else:
        return "Fall"

df["season"] = df["month"].apply(get_season)
seasons_counts = df["season"].value_counts()
seasons_counts
```

```python
plt.figure(figsize=(10,8))
months_count.plot(kind='bar' , color="skyblue")
plt.title("Courses Published by Month" , fontsize=15)
plt.xlabel("Month" , fontsize=12)
plt.ylabel("NumberOfCourses" , fontsize=12)
plt.xticks(rotation=45 , ha='right')
plt.show()
```

```python
plt.figure(figsize=(10,8))
seasons_counts.plot(kind="bar" , color="skyblue")
plt.title("Courses Published by Seasons" , fontsize=15)
plt.xlabel("Season" ,fontsize=12)
plt.ylabel("NumberOfCourses" , fontsize=12)
plt.xticks(rotation=45 , ha="right")
plt.show()
```
